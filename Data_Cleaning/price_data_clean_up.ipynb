```python
import pandas as pd
import datetime as dt
```

```python
price_index=pd.read_csv("../Resources/New_housing_index_price.csv")
price_index.head()
```

```python
price_index = price_index.drop(["TERMINATED", "SYMBOL","DECIMALS", "DGUID","STATUS","UOM","UOM_ID","SCALAR_FACTOR","SCALAR_ID","VECTOR","COORDINATE"], axis=1)
```

```python
price_index.head()
```

```python
price_index["REF_DATE"].astype("string")
```

```python
price_index = price_index.rename(columns = {"VALUE":"PRICE_INDEX","New housing price indexes":"Housing_Type"})
```

```python
price_index.head()
```

```python
price_index=price_index.dropna()
```

```python
price_index=price_index[price_index.Housing_Type != "Total (house and land)"]
price_index.head()
```

```python
price_index["GEO"].nunique()
```

```python
price_index=price_index[price_index.GEO!="Census metropolitan areas"]
```

```python
price_index["GEO"].astype("str")
```

```python
price_index["PRICE_INDEX"].astype("int")
```

```python
df=price_index.GEO.str.split(", ",expand=True,)
df.head()
```

```python
df_2=pd.concat([price_index,df], axis=1)
```

```python
df_2.columns.tolist()
```

```python
df_2 = df_2.rename(columns = {0:"City",1:"Province"})
df_2.head()
```

```python
df_2.to_csv("../Resources/price_index_clean.csv",index=True)
```

```python
df_2=df_2.drop(columns={2,3,'GEO'})
df_2.head()
```

```python
df_2=df_2[df_2.Province.notnull()]
df_2.head()
```

```python
cols=['REF_DATE','City','Province','Housing_Type','PRICE_INDEX']
df_2=df_2[cols]
df_2.head()
```

```python
df_2.index.names=["housingprice_index"]
```

```python
df_2[["Ref_Year","Ref_Month","Day"]]=df_2.REF_DATE.str.split("-",expand=True)
df_2.head()
```

```python
df_2=df_2.drop(["REF_DATE","Day"],axis=1)
cols=['Ref_Year','Ref_Month','City','Province','Housing_Type','PRICE_INDEX']
df_2=df_2[cols]
df_2.head()
```
