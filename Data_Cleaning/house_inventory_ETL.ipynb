```python
import pandas as pd
```

```python
house_inventory_df=pd.read_csv('../Resources/new_housing_inventory.csv')
house_inventory_df.head()
```

```python
house_inventory_df.dtypes
```

```python
housing_inventory = house_inventory_df.drop(["TERMINATED", "SYMBOL","DECIMALS", "DGUID","STATUS","UOM","UOM_ID","SCALAR_FACTOR","SCALAR_ID","VECTOR","COORDINATE"], axis=1)
housing_inventory.head()
```

```python
housing_inventory["REF_DATE"].astype("datetime64")
```

```python
housing_inventory["Year"] = pd.DatetimeIndex(housing_inventory['REF_DATE']).year
housing_inventory["Month"] = pd.DatetimeIndex(housing_inventory['REF_DATE']).month
housing_inventory.head()
```

```python
housing_inventory = housing_inventory.rename(columns = {"VALUE":"UNIT_VALUE","Type of dwelling unit":"Dwelling_Type","Completed dwelling units":"Completed_units"})
housing_inventory.head()
```

```python
housing_inventory=housing_inventory.dropna()
```

```python
housing_inventory=housing_inventory[housing_inventory.Dwelling_Type!="Total units"]
```

```python
housing_inventory=housing_inventory[housing_inventory.GEO!="Census metropolitan areas"]
housing_inventory.head()
```

```python
housing_inventory.reset_index()
```

```python
housing_inventory["GEO"].astype("str")
```

```python
df=housing_inventory.GEO.str.split(", ",expand=True,)
df.head()
```

```python
df_2=pd.concat([housing_inventory,df], axis=1)
df_2 = df_2.rename(columns = {0:"City",1:"Province"})
df_2=df_2.drop(columns={2,'GEO'})
cols=['REF_DATE','City','Province','Completed_units','Dwelling_Type','UNIT_VALUE']
df_2=df_2[cols]
df_2=df_2.reset_index(drop=True)
# df_2.index.names=["inventory_index"]
df_2.head()
```

```python
# df = df.pivot_table(
#     values='value',
#     index=['stream_name', 'preferred_timestamp', 'internal_timestamp'],
#     columns='value_id'
#     )
```

```python
df_2=df_2.drop('Province', axis=1)
df_2
```

```python
df_3=df_2.pivot_table(
    values='UNIT_VALUE',
    index=['REF_DATE','Completed_units','Dwelling_Type'],
    columns='City')
df_3.head()
```

```python
df_3.reset_index(inplace=True)
df_3.columns.name = None
df_3.head()
```

```python
df_3.to_csv("../Resources/ml_housing_inventory_clean.csv",index=True)
```
