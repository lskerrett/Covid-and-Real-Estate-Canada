```python
import pandas as pd
```

```python
city_df = pd.read_csv("../Resources/canada_pop_size_2011_2016.csv")
city_df.head()
```

```python
city_df.columns.tolist()
```

```python
city_df.dtypes
```

```python
city_df = city_df.drop(["Geographic code", "Geographic name, french","Population centre size group, french", "Type of population centre, french","Incompletely enumerated Indian reserves and Indian settlements, 2016", "Province / territory, french", "Additional province code", "2011 adjusted population flag", "Population, % change", "Total private dwellings, 2016", "Private dwellings occupied by usual residents, 2016", "Geographic code, Province / territory", "Population, 2011"], axis=1)
city_df.head()
```

```python
city_df = city_df.rename(columns= {"Geographic name, english" : "city_name", "Population centre size group, english": "city_size_type", "Type of population centre, english": "city_type", "Province / territory, english": "province", "Population, 2016": "pop_2016", "Population, 2011": "pop_2011", "Land area in square kilometres, 2016": "city_size", "Population density per square kilometre, 2016": "pop_density"})
city_df.head()
```

```python
city_df = city_df.dropna()
```

```python
city_df.to_csv("../Resources/cleaned_canada_pop_2011_2016.csv", index=False)
```
